```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [10,20,30,40,50]
```

```python
plt.plot(x, y ,marker='o',linestyle ='-',color='b',label='Tred')
plt.xlabel('x_data')
plt.ylabel('y_data')
plt.title('line plot graph')
plt.legend()
plt.show()
```

# Bar Graph

```python
cat = ['A','B','C','D']
values = [30,50,20,80]
```

```python
plt.bar(cat, values, color =['blue', 'green', 'red', 'purple'])
plt.xlabel('categories')
plt.ylabel('values')
plt.title('Bar chart with categories and values')
plt.show()
```

```python
sns.barplot(x=cat, y=values)
plt.show()
```

# Histogram

```python
import numpy as np
```

```python
data = np.random.random(1000)
```

```python
data
```

```python
plt.hist(data,bins=30, color ='skyblue',edgecolor ='black')
plt.xlabel('Values on X')
plt.ylabel('Values on Y')
plt.title('Histogram plot')
plt.show()
```

# Box Plot

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
sns.boxplot(x =df['Fare'])
plt.title('Box plot on fare')
plt.show()
```

# co-relation

```python
cr_data = df[['Survived','Pclass','Age','Fare']]
```

```python
corelation = cr_data.corr()
sns.heatmap(corelation, annot=True, cmap='coolwarm')
plt.title('HeatMap')
plt.show()
```
